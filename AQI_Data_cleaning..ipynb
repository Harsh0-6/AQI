# Data Cleaning and Processing
After analysis, we can select only necessary columns (which is calculated by most of the cities in India).

1. City is removed since it's not necessary.
2. Date is removed since AQI is not dependent on date.
3. NO and Nox is removed since we have considered No2 which is subsidary os Nox.
4. Benzene, Toluene, Xylene is removed since only major cities consider these paramater while calculating AQI while our model is for whole country so removing them will be better option for any type of city person.
5. AQI Bucket is removed since it is categorical data and AQI Bucket can be calculated using AQI values by simple formula.
6. The reason for dropping PM10 is because we are considering PM2.5, so PM2.5 and PM10 contribute almost the same data, so considering both will make model too dependent on these features.
7. Also PM10 has high NaN values(almost 35% of the data is NaN).
8. Dropping one or the other will not affect the AQI since PM2.5 and PM10 delivers almost same information.
9. Since VIF value of PM10 is high and To avoid multicollinerity problem, we can drop PM10.

```python
#Removing unneccesary columns.
df.drop(["City", "PM10", "Date","NO","NOx","NH3", "Benzene", "Toluene", "Xylene","AQI_Bucket"], axis=1, inplace=True)
```
